```python
import zipfile, os

ZIP_PATH = "Assignment 9.v3i.yolov11.zip"  # or use files.upload() if not in content
EXTRACT_DIR = "/content/Assignment_9_extracted"
os.makedirs(EXTRACT_DIR, exist_ok=True)

with zipfile.ZipFile(ZIP_PATH, 'r') as z:
    z.extractall(EXTRACT_DIR)

print("âœ… Extracted successfully to:", EXTRACT_DIR)
```

```python
import os

EXTRACT_DIR = "/content/Assignment_9_extracted"

image_dirs = set()
for root, dirs, files in os.walk(EXTRACT_DIR):
    for f in files:
        if f.lower().endswith((".jpg", ".jpeg", ".png")):
            image_dirs.add(root)
image_dirs = sorted(list(image_dirs))
print("ğŸ–¼ï¸ Found image folders:\n", "\n".join(image_dirs))

if image_dirs:
    IMAGE_DIR = image_dirs[0]  # use the first found folder automatically
    print("\nâœ… Using image directory for detection:\n", IMAGE_DIR)
else:
    raise FileNotFoundError("No image folders found inside extracted directory.")
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import os

# Load your trained YOLOv8 model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Use the test image folder
IMAGE_DIR = "/content/Assignment_9_extracted/test/images"

# Run detection and save annotated results
results = model.predict(source=IMAGE_DIR, conf=0.4, save=True)
output_dir = results[0].save_dir
print("âœ… Detection images saved in:", output_dir)

# Display one detected image inline
first_image = None
for f in os.listdir(output_dir):
    if f.lower().endswith(('.jpg', '.jpeg', '.png')):
        first_image = os.path.join(output_dir, f)
        break

if first_image:
    img = cv2.imread(first_image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,6))
    plt.imshow(img)
    plt.axis("off")
    plt.title("âœ… License Plate Detected Successfully!")
    plt.show()
else:
    print("âš ï¸ No annotated image found in:", output_dir)
```
